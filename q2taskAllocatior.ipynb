```python
from datetime import datetime

def AllocateTask():
    '''  Prompts the user to enter a start date and end date in yymmdd format and handles invalid inputs.
    
        Then asks the user to enter the date to allocate tasks , if it's a weekend it won't allocate tasks.

        Returns:
        A dictionary with task allocations as per date 
    '''
    try:
        
        startDate = input("Enter start date in yymmdd format: ")
        endDate = input("Enter end date in yymmdd format: ")

        
        start=datetime.strptime(startDate, '%y%m%d')
        end=datetime.strptime(endDate, '%y%m%d')

        tasks= [i for i in range(int(start.strftime("%y%m%d")), int(end.strftime("%y%m%d")) + 1)]
        allocations= {task: "" for task in tasks}

        while True:
            allocateTask= input("Enter date to allocate task for in yymmdd format ONLY: ")

            try:
                allocateTaskInt= int(allocateTask)
                if allocateTaskInt in allocations:
                   
                    weekday= datetime.strptime(allocateTask, "%y%m%d").isoweekday()

                    
                    if weekday in [6, 7]:
                        print("It's a weekend (Saturday or Sunday), so no tasks can be assigned")
                        add_more = input("Do you want to add tasks for another date? yes/no : ").strip().lower()
                        if add_more != 'yes':
                            break
                    else:
                        
                        new_task= input("Enter the task: ")
                        allocations[allocateTaskInt] += ", " +new_task
                        add_more= input("Do you want to add another task for a different date? yes/no: ").strip().lower()
                        if add_more!= 'yes':
                            break
                else:
                    print("Invalid date entered. Please try again.")
            except ValueError:
                print("Invalid date format. Please enter the date in yymmdd format ONLY")
        
        return allocations

    except ValueError as e:
        print(f"Error: {e}. Please enter the date in the correct format yymmdd")
        return {}


allocations= AllocateTask()
print(allocations)
```
